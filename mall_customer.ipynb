```python
x=data.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
k =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init = 'k-means++',random_state =0)
    kmeans.fit(x)
    k.append(kmeans.inertia_)
    
plt.plot(range(1,11),k,marker='o')    
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(n_clusters=5,init = 'k-means++',random_state =0)
kmeans=model.fit_predict(x)
      
```

```python
cluster_centers = model.cluster_centers_
print("Cluster Centers: \n", cluster_centers)

# Visualize the clusters (for 2D data)

plt.scatter(x[:, 0], x[:, 1], c=kmeans, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='Black', marker='o', s=200, label='Centroids')
plt.title("KMeans Clustering")
plt.legend()
plt.show()

#  Evaluate clustering quality with silhouette score
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(x, kmeans)
print("Silhouette Score:", silhouette_avg)
```
